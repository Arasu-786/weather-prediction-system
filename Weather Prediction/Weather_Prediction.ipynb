```python
#Load Dataset
import pandas as pd
df = pd.read_csv('/content/weatherAUS.csv')
df.head()
```

```python
#Finding Missing Values in the Dataset
df.info()
df.isnull().sum()
```

```python
#Data Cleaning
df = df.dropna()
```

```python
#Converting Output
df['RainTomorrow'] = df['RainTomorrow'].map({'Yes': 1, 'No': 0})
```

```python
#Adding Features
features = ['MinTemp', 'MaxTemp', 'Humidity9am', 'Humidity3pm', 'WindSpeed9am']

X = df[features]
y = df['RainTomorrow']
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Model Training
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
#Accuracy Checking
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#Prediction with Sample
sample = [[20, 25, 90, 85, 10]]

prediction = model.predict(sample)

if prediction[0] == 1:
    print("Rain Tomorrow 🌧️")
else:
    print("No Rain ☀️")
```

```python
#Weather Prediction Model
min_temp = float(input("Min Temperature: "))
max_temp = float(input("Max Temperature: "))
hum9 = float(input("Humidity at 9am: "))
hum3 = float(input("Humidity at 3pm: "))
wind = float(input("Wind Speed at 9am: "))

sample = [[min_temp, max_temp, hum9, hum3, wind]]

prediction = model.predict(sample)

if prediction[0] == 1:
    print("Rain Tomorrow 🌧️")
else:
    print("No Rain ☀️")
```
